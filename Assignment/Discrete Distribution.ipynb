POISSON DISTRIBUTION

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import poisson 

#set the avg. rate per interval 
lambda_parameter=3

random_sample=poisson.rvs(mu=lambda_parameter,size=1000)

plt.hist(random_sample,bins=np.arange(0,max(random_sample)+1.5)-0.5,density=True,alpha=0.75)

#plot the probablity mass function of the poission distrubution 
x=np.arange(0,max(random_sample)+1)
pmf_values=poisson.pmf(x,mu=lambda_parameter)
plt.vlines(x,0,pmf_values,colors='red',lw=2,alpha=0.75)

plt.xlabel('number of events')
plt.ylabel('probablity')
plt.title(f'poisson distribution(Î»={lambda_parameter})')

#show plot 
plt.show()
```

Example=1 The average rate of a car accident on a perticular stretch of road is 0.5 accidents per day,what is the probablity of observing 2 or more accidents on a given day 

```python
from scipy.stats import poisson 
lambda_functions=0.5 
k_accidents_2_or_more=2 
prob_2_more_accidents=1-poisson.cdf(k_accidents_2_or_more-1,lambda_functions)
print(f"problem 1:probablity of 2 or more accidents in one day:{prob_2_more_accidents}")
```

example-2 in a software development project ,on avg ,3 bugs are reported per day ,what is the probablity that there are no bugs reported on a given day

```python
from scipy.stats import poisson 
lambda_bugs=3 
no_of_bugs=0 
bugs=poisson.pmf(no_of_bugs,lambda_bugs)
print(f"problem 1:probablity of no bugs reported on a  day:{bugs}")
```

Geometric Distrubution 

```python
from scipy.stats import geom
```

```python
#set the probablity 
p=0.3 
#create n array of possible values for the number of trials
t=np.arange(1,11)
#calculate the probablity mass function for each values 
pmf_val=geom.pmf(t,p)
print("pmf_val:")

for i,pmf in zip(t,pmf_val):
    print(f"p(x={i})={pmf:.4f}") #.4f upto 4 decimal places 


#plot the pmf 

plt.bar(t,pmf_val,align='center',alpha=0.7)
plt.xlabel('number of trials(i)')
plt.ylabel('probablity')
plt.title(f'geometric distribution(p={p})')
```
