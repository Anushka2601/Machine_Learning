```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import norm
```

```python
#define the parameters of the normal distribution 
mean=0 
std_dev=1 

#generate a range of x values 
x=np.linspace(-5,5,1000)

#calculate the probablity density function for each value 
pdf_val=norm.pdf(x,mean,std_dev)

#plot the normal distribution 
plt.plot(x,pdf_val,label='Normal Distribution')
plt.title('Normal Distribution')
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.legend() #right upper side normal distribution 
plt.show()
```

```python
std=5 
mean=60
lower=norm.ppf(0.05,mean,std)
#print(lower)
upper=norm.ppf(0.95,mean,std)



x_values = np.linspace(mean - 4* std, mean + 4 * std, 1000)

# Calculate PDF values
pdf_values = norm.pdf(x_values, mean,std)
plt.plot(x_values, pdf_values, label='Normal Distribution')


plt.fill_between(x_values, pdf_values, where=((x_values > lower) & (x_values < upper)), color='red', alpha=0.5, label='Middle 90%')

# Mark the separating values
plt.axvline(x=lower, color='green', linestyle='--', label='Lower Bound')
plt.axvline(x=upper, color='blue', linestyle='--', label='Upper Bound')

# Add labels and legend
plt.title('Separating the Middle 90% of a Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()

# Show the plot
plt.show()
```

```python
#Import Required Libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Set the seed for reproducibility
np.random.seed(42)

# Parameters for the exponential distribution
lambda_param = 0.5  # decay constant (rate parameter)

# Generate random numbers from an exponential distribution
data = np.random.exponential(scale=1/lambda_param, size=1000)

# Plot the histogram of the generated data
plt.hist(data, bins=30, density=True, alpha=0.7, color='pink', label='Histogram')

# Plot the probability density function (PDF) of the exponential distribution
x = np.linspace(0, 10, 1000)
pdf = expon.pdf(x, scale=1/lambda_param)
plt.plot(x, pdf, 'g-', lw=2, label='Exponential PDF')

# Add labels and a legend
plt.title('Exponential Distribution Demonstration')
plt.xlabel('Time')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()

# Show the plot
plt.show()
```

```python
#Import Required Library
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon  # Exponential distribution for illustration

# Parameters for the distribution
scale_parameter = 1.0

# Generate values for time (t)
t_values = np.linspace(0, 5, 1000)

# Calculate survival function (SF)
sf_values = expon.sf(t_values, scale=scale_parameter)

# Calculate cumulative distribution function (CDF)
cdf_values = expon.cdf(t_values, scale=scale_parameter)

# Plot the survival function and cumulative distribution function
plt.figure(figsize=(10, 6))
plt.plot(t_values, sf_values, label='Survival Function (SF)', color='blue')
plt.plot(t_values, cdf_values, label='Cumulative Distribution Function (CDF)', color='orange')
plt.title('Survival Function vs. Cumulative Distribution Function')
plt.xlabel('Time (t)')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```
