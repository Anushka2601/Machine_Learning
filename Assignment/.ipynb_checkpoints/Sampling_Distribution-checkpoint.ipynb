```python
#Import Required Libraries
import numpy as np
import matplotlib.pyplot as plt

# Population parameters
population_mean = 75
population_std_dev = 10
population_size = 1000

# Sample parameters
sample_size = 30
num_samples = 1000

# Step 2 and 3: Generate multiple samples and calculate means
sample_means = [np.mean(np.random.normal(population_mean, population_std_dev, sample_size)) for _ in range(num_samples)]

# Step 4: Calculate Standard Error
standard_error = population_std_dev / np.sqrt(sample_size)

# Step 5: Plot the sampling distribution
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='blue', label='Sampling Distribution')
plt.title('Sampling Distribution of the Mean')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

# Plot a normal distribution for comparison
normal_distribution = np.random.normal(population_mean, standard_error, 10000)
plt.hist(normal_distribution, bins=30, density=True, alpha=0.5, color='red', label='Normal Distribution')

plt.legend()
plt.show()
```

Suppose a university claims that the average time its students take to complete a certain exam is 60 minutes. However, you believe that the average time is actually less than 60 minutes. You decide to conduct a left-tailed t-test to test your hypothesis.

```python
from scipy.stats import stats 
import stats 
```

```python
sample_mean = 58
population_mean = 60
sample_std_dev = 8
sample_size = 20

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))

# Degrees of freedom
degrees_of_freedom = sample_size - 1

# Significance level
alpha = 0.05

# Critical value from t-distribution
critical_value = stats.t.ppf(alpha, degrees_of_freedom)

# p-value from t-test
p_value = stats.t.cdf(t_statistic, degrees_of_freedom)

# Make a decision based on the test result
if t_statistic < critical_value:
    print("Reject the null hypothesis. There is evidence to suggest that the average time is less than 60 minutes.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to suggest that the average time is less than 60 minutes.")
```
