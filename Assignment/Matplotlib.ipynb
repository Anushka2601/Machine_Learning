```python
student=[8,5,10,12,9]
marks=[45,57,89,12,85]

plt.plot(student,marks)
plt.title("StudentsMarks")
plt.xlabel("students")
plt.ylabel("marks")
```

Bar Chart

```python
l=[78,89,45,63,12]
l2=[4,5,6,8,9]

plt.bar(l,l2)
plt.title("barchart")
plt.xlabel("Marks")
plt.ylabel("No of Students")
plt.show()
```

BarChart with color change

```python
marks=[54,89,74,78,89]
students=[4,2,1,0,8]
plt.bar(marks,students,color='g')
plt.title("StudentPerformance")
plt.xlabel("Marks")
plt.ylabel("StudentNumber")
plt.show()
```
